```python
#in max scaling


mn = df['Total Marks'].min()
mx = df['Total Marks'].max()

df['Scaled Marks'] = df['Total Marks'].apply(lambda x: (x-mn)/(mx-mn))

df
```

```python
def grading_system(marks):
  if marks>= 260:
    return 'A+'
  elif marks>= 240:
    return 'A'
  elif marks>= 220:
    return 'A-'

df['Grade'] = df['Total Marks'].apply(grading_system)

df
```

```python
from datetime import datetime, timedelta
import random

#     
names = [
    "Rahim Ahmed", "Karim Hossain", "Sumaiya Akter", "Nusrat Jahan", "Hasan Ali",
    "Aminul Islam", "Rafi Chowdhury", "Tania Rahman", "Sadia Haque", "Mamun Khan",
    "Fahim Uddin", "Rima Sultana", "Zahid Hasan", "Shorna Das", "Imran Hossain",
    "Tahsin Alam", "Rubel Mia", "Mitu Akter", "Asif Rahman", "Lamia Noor"
]

#   
instructors = ["Dr. Mahbub", "Ms. Farhana", "Mr. Rashed", "Dr. Kabir"]
locations = ["Dhaka", "Chittagong", "Rajshahi", "Sylhet"]

#   
data = []
for i in range(1, 21):
    enroll_date = datetime(2024, random.randint(1, 6), random.randint(1, 28))
    finish_date = enroll_date + timedelta(days=random.randint(30, 120))
    status = random.choice(["Completed", "Ongoing", "Dropped"])
    total_marks = np.random.randint(50, 100) if status == "Completed" else np.nan

    data.append({
        "student_id": f"S{i:03d}",
        "full_name": names[i-1],
        "completion_status": status,
        "enrollment_date": enroll_date.strftime("%Y-%m-%d"),
        "finish_date": finish_date.strftime("%Y-%m-%d"),
        "instructor": random.choice(instructors),
        "location": random.choice(locations),
        "total_marks": total_marks
    })

# DataFrame 
df = pd.DataFrame(data)

df
```

```python
df['EnrollmentDate'] = pd.to_datetime(df['enrollment_date'])
```

```python
df['EnrollmentDate']
```

```python
df['Enrollment Year'] = df['EnrollmentDate'].dt.year
df['Enrollment Date'] = df['EnrollmentDate'].dt.day
```

```python
df
```

```python
df['FinishedDate'] = pd.to_datetime(df['finish_date'])

df['Total time taken to finish']= df['FinishedDate'] - df['EnrollmentDate']

df
```
